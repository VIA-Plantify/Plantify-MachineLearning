```python
import kagglehub
import shutil
import os

# download dataset
path = kagglehub.dataset_download("miadul/irrigation-water-requirement-prediction-dataset")
path2 = kagglehub.dataset_download("ziya07/plant-health-data")



# current folder (where your script runs)
current_dir = os.getcwd()

# copy everything from dataset folder into current folder
for item in os.listdir(path):
    src = os.path.join(path, item)
    dst = os.path.join(current_dir, item)

    if os.path.isdir(src):
        shutil.copytree(src, dst, dirs_exist_ok=True)
    else:
        shutil.copy2(src, dst)
for item in os.listdir(path2):
    src = os.path.join(path2, item)
    dst = os.path.join(current_dir, item)

    if os.path.isdir(src):
        shutil.copytree(src, dst, dirs_exist_ok=True)
    else:
        shutil.copy2(src, dst)

print("Copied dataset into current folder:", current_dir)
```
